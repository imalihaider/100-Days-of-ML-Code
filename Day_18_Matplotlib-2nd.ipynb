```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a scatter plot with a colormap
plt.scatter(iris['sepal_length'], iris['sepal_width'], c=iris['petal_length'], cmap='viridis')
plt.colorbar()
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Sepal Length vs Width')
plt.show()
```

```python
# Create a histogram with multiple subplots
fig, axs = plt.subplots(1, 3, figsize=(10, 4), sharey=True)
sns.histplot(iris['sepal_length'], kde=False, color='b', ax=axs[0])
axs[0].set_xlabel('Sepal Length')
axs[0].set_ylabel('Count')
axs[0].set_title('Iris Sepal Length')

sns.histplot(iris['sepal_width'], kde=False, color='g', ax=axs[1])
axs[1].set_xlabel('Sepal Width')
axs[1].set_title('Iris Sepal Width')

sns.histplot(iris['petal_length'], kde=False, color='r', ax=axs[2])
axs[2].set_xlabel('Petal Length')
axs[2].set_title('Iris Petal Length')

fig.suptitle('Histogram of Iris Features')
plt.show()
```

```python
# Group the data by species and calculate the means
means = iris.groupby('species').mean()

# Create a stacked bar chart
means.plot(kind='bar', stacked=True)
plt.xlabel('Species')
plt.ylabel('Feature Value')
plt.title('Mean Feature Values by Iris Species')
plt.show()
```
