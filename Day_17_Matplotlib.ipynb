```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
# Load the Iris dataset
iris = load_iris()
# Get the data and target values
X = iris.data
y = iris.target
# Set the color for each target class
colors = ['r', 'g', 'b']
# Add labels and a title
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Iris dataset')
```

```python
# Scatter plot with color-coded species
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.show()
```

```python
# Histograms of sepal length for each species
sns.histplot(data=iris, x="sepal_length", hue="species", multiple="stack")
plt.show()
```

```python
# Boxplot of petal width by species
sns.boxplot(x="species", y="petal_width", data=iris)
plt.show()
```

```python
# Violin plot of petal length by species
sns.violinplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
# Some more plots as for understanding without using Iris Datasets
# Sample data
labels = ['Apples', 'Oranges', 'Bananas', 'Grapes']
sizes = [23, 17, 35, 29]

# Creating a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Setting the title
plt.title('Fruit Quantity Distribution')

# Displaying the plot
plt.show()
```
